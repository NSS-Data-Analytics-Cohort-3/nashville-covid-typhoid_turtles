```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import FloatImage
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import time
import numpy as np
```

```python
cluster_frame = pd.read_csv('../data/violation_distance.csv')
cluster_frame.head()
```

```python
geometry = [Point(xy) for xy in zip(cluster_frame.Longitude, cluster_frame.Latitude)]
# cluster_frame = cluster_frame.drop(['Latitude', 'Longitude'], axis=1)
cluster_geo = gpd.GeoDataFrame(cluster_frame, crs="EPSG:4326", geometry=geometry)
```

```python
cluster_map = folium.Map(location = [36.16713345170994, -86.76850874477554], zoom_start = 12)
folium.TileLayer('stamentoner').add_to(cluster_map)
```

```python
cluster_frame.head()
```

```python
for row_index, row_values in cluster_geo.iterrows():
    loc = [row_values['Latitude'], row_values['Longitude']]
    pop = 'Outbreak cluster: ' + row_values['Cluster Name']
    marker = folium.Marker(
        location = loc,
        icon = folium.Icon(color="red",icon="medkit", prefix='fa'),
#        radius=100,
        popup = pop,
        fill=True,
        color='red') 
    
    marker.add_to(cluster_map)
```

```python
for row_index, row_values in cluster_geo.iterrows():
    loc = [row_values['violation_lat'], row_values['violation_lon']]
    marker = folium.Circle(
        location = loc,
        radius=200,
        fill=True,
        color='blue') 
    
    marker.add_to(cluster_map)

url = ('https://i.imgur.com/YUynSOr.png')
folium.plugins.FloatImage(url, bottom=75, left=80).add_to(cluster_map)
cluster_map.save('../maps/outbreak_violations_map.html')
cluster_map
```
